```python
import pandas as pd
import numpy as np
data = pd.read_csv('./data/data.csv')
data.head()
```

```python
from matplotlib import pyplot as plt
fig1 = plt.figure()
plt.scatter(data.loc[:,'X1'],data.loc[:,'X2'])
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
X = data.drop(['label'], axis=1)
y = data.loc[:, 'label']

pd.value_counts(y)
```

```python
fig1 = plt.figure()
label0 = plt.scatter(data.loc[:,'X1'][y==0],data.loc[:,'X2'][y==0])
label1 = plt.scatter(data.loc[:,'X1'][y==1],data.loc[:,'X2'][y==1])
label2 = plt.scatter(data.loc[:,'X1'][y==2],data.loc[:,'X2'][y==2])

plt.title("labeled data")
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```

```python
#establish a KNN model
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(X,y)
```

```python
y_predict_knn = KNN.predict(X)
```

```python
fig6 = plt.subplot(121)
label0 = plt.scatter(X.loc[:,'X1'][y_predict_knn==0],X.loc[:,'X2'][y_predict_knn==0])
label1 = plt.scatter(X.loc[:,'X1'][y_predict_knn==1],X.loc[:,'X2'][y_predict_knn==1])
label2 = plt.scatter(X.loc[:,'X1'][y_predict_knn==2],X.loc[:,'X2'][y_predict_knn==2])

plt.title("knn results")
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend((label0,label1,label2),('label0','label1','label2'))

fig6 = plt.subplot(122)
label0 = plt.scatter(X.loc[:,'X1'][y==0],X.loc[:,'X2'][y==0])
label1 = plt.scatter(X.loc[:,'X1'][y==1],X.loc[:,'X2'][y==1])
label2 = plt.scatter(X.loc[:,'X1'][y==2],X.loc[:,'X2'][y==2])

plt.title("original data")
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend((label0,label1,label2),('label0','label1','label2'))
plt.show()
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y,y_predict_knn))
```
